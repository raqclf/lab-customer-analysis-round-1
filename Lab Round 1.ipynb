```python
#Rearrange the columns in the dataframe as needed and Standardize header names.
file1 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})
file2 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})
file3 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})
```

```python
file2 = file2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
file3 = file3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#Concatenate the three dataframes
data = pd.concat([file1, file2, file3], axis=0)
data.shape
```

```python
#Which columns are categorical?
data.select_dtypes(['object'])
```

```python
#Delete the column education and the number of open complaints from the dataframe.
data = data.drop(['Education', 'Number of Open Complaints'], axis = 1)
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
data1 = data.copy()

listdata = []
listdataclean = []

for value in data1:
    listdata.append(str(value))
    
for value in listdata:
    listdataclean.append(value.replace('%',''))

listdataclean = pd.to_numeric(listdataclean, errors='coerce')
                              
data1['Customer Lifetime Value'] = listdataclean/100
```
